# Connecting to Cassandra from Python

First we have to install Cassandra Driver from DataStax. 

pip install cassandra-driver

We can now import cluster which allows us to connect to our key space then our table. 

```python
from cassandra.cluster import Cluster
import pandas as pd
```

```python
cluster = Cluster()
session = cluster.connect('netflix')
session.execute("INSERT INTO emp (emp_id, emp_name, emp_city, emp_phone, emp_sal) VALUES(1,'ram', 'Hyderabad', 9848022338, 50000)")
```

```python
def pandas_factory(colnames, rows):
    return pd.DataFrame(rows, columns=colnames)
session.row_factory = pandas_factory
session.default_fetch_size = None
query = "SELECT * from netflix"
rslt = session.execute(query, timeout=None)
df1 = rslt._current_rows
```
